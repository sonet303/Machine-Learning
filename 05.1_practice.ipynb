```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_linear = np.linspace(1, 10, 50)
y_linear = 2 * x_linear + 5 + np.random.randn(50) * 2
```

```python
x_linear.shape
```

```python
x_nonlinear = np.linspace(1, 10, 50)
y_nonlinear = 2 * np.sin(x_nonlinear) + np.random.randn(50)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(x_linear, y_linear, color = 'green', label = 'Linear Data')
plt.plot(x_linear, 2*x_linear+5, color = 'red', label = 'True Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Data')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x_nonlinear, y_nonlinear, color = 'green', label = 'NonLinear Data')
plt.plot(x_linear, 2*np.sin(x_nonlinear), color = 'red', label = 'True Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('NonLinear Data')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
correlation_Linear = np.corrcoef(x_linear, y_linear)[0, 1]
correlation_NonLinear = np.corrcoef(x_nonlinear, y_nonlinear)[0, 1]
print(correlation_Linear)
print(correlation_NonLinear)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg1 = LinearRegression()
reg1.fit(x_linear.reshape(-1, 1), y_linear)
```

```python
reg1.score(x_linear.reshape(-1, 1), y_linear)
```

```python
plt.plot(x_linear, reg1.predict(x_linear.reshape(-1, 1)), label = 'Predicted Line')
plt.scatter(x_linear, y_linear, color = 'green', label = 'Linear Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Data')
plt.legend()
```

```python
reg1.predict([[1.5]])
```

# Non Linear

```python
reg2 = LinearRegression()
```

```python
reg2.fit(x_nonlinear.reshape(-1,1), y_nonlinear)
```

```python
reg2.score(x_nonlinear.reshape(-1,1), y_nonlinear)
```

```python
plt.plot(x_nonlinear, reg2.predict(x_nonlinear.reshape(-1,1)))
```

```python
plt.scatter(x_nonlinear, y_nonlinear, color = 'green', label = 'NonLinear Data')
plt.plot(x_nonlinear, reg2.predict(x_nonlinear.reshape(-1,1)), label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('NonLinear Data')
plt.legend()
```
