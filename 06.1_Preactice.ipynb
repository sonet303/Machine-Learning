```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df1 = pd.read_csv('linear_data.csv')
df1.head()
```

```python
df1.shape
```

```python
x_linear = df1[['x']]
y_linear = df1['y']
```

```python
plt.scatter(x_linear, y_linear)
```

```python
df1.corr()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg1 = LinearRegression()
```

```python
reg1.fit(x_linear, y_linear)
```

```python
reg1.score(x_linear, y_linear)
```

```python
plt.scatter(x_linear, y_linear)
plt.plot(x_linear, reg1.predict(x_linear), color = 'yellow')
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=6)
```

```python
X_poly = poly.fit_transform(x_nonlinear)
```

```python
X_poly.shape
```

```python
reg_poly = LinearRegression()
```

```python
reg_poly.fit(X_poly, y_nonlinear)
```

```python
reg_poly.score(X_poly, y_nonlinear)
```

```python
plt.scatter(x_nonlinear, y_nonlinear, color='green', label='Data with Non-linear Relationship')
plt.plot(x_nonlinear, 2 * np.sin(x_nonlinear), color='orange', label='True Non-linear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-linear Relationship')
plt.legend()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_poly, y_nonlinear, test_size =.30, random_state = 42)
```

```python
reg = LinearRegression()
```

```python
reg.fit(xtrain, ytrain)
```

```python
reg.score(X_poly, y_nonlinear)
```

```python
reg.score(xtrain, ytrain)
```

```python
reg.score(xtest, ytest)
```
